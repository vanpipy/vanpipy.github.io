# The Concept Of A Limit

Let us consider a function $f(x)$ that is defined for all values of $x$ near a point $a$ on the x-axis but not necessarily at $a$ itself. Suppose there exists a real number $L$ with the property that $f(x)$ gets closer to $L$ as $x$ near a point closer and closer to $a$. Under these circumstances we say that $L$ is the __limit__ of $f(x)$ as $x$  approaches $a$, and we express this symbolically by writing

$$lim_{x \to a} f(x) = L$$
$$|x - a| < \delta$$
$$|f(x) - L| < \epsilon$$

when $lim_{x \to a} f(x) = L$ and $lim_{x \to a} g(x) = M$, then

$$lim_{x \to a}[f(x) + g(x)] = L + M$$
$$lim_{x \to a}[f(x) - g(x)] = L - M$$
$$lim_{x \to a}[f(x) \cdot g(x)] = L \cdot M$$
$$lim_{x \to a}[f(x) / g(x)] = L / M$$

With a sight chance of notation, we can express the continuity of our function $f(x)$ at a point $x$ (instead of $a$) in either of the equivalent forms

$$lim_{\Delta{x} \to 0} f(x + \Delta{x}) = f(x)$$

or

$$lim_{\Delta{x} \to 0} [f(x + \Delta{x}) - f(x)] = 0$$

## The Mean Value Theorem

If $f(x)$ is continous on the closed interval $[a, b]$; and

$f(x)$ is differentiable on the open interval $(a, b)$.

Then there exists at least one point $c$ in the open interval $(a, b)$ such that

$$f'(x) = \frac{f(b) - f(a)}{b - a}$$

```python
import matplotlib.pyplot as plt
import numpy as np

def f(x):
    return x**3 - 3*x**2

def df(x):
    return 3*x**2 - 6*x

# Interval [a, b]
a, b = 0, 2
c = 0.42

x = np.linspace(a-0.5, b+0.5, 400)
plt.figure(figsize=(10,6))

plt.plot(x, f(x), label='f(x)', lw=2)
plt.plot([a, b], [f(a), f(b)], '--', label='Secant line')
plt.plot(x, f(c) + df(c)*(x - c), ':', label=f'Tangent at c={c}')

plt.scatter([a, b], [f(a), f(b)], color='black')
plt.title('Mean Value Theorem')
plt.legend()
plt.grid(True)
plt.show()
```

## The Extreme Value Theorem

If $y = f(x)$ is a function that is defined and continuous on a closed interval $[a, b]$, then this function attains both a maximum value and a minimum value at points of the intervals; that is, there exist points $c$ and $d$ in $[a, b]$ such that $f(c) \geq f(x) \geq f(d)$ for all $x$ in $[a, b]$.


```python
import matplotlib.pyplot as plt
import numpy as np

def f(x):
    return 1/x

# Interval [a, b]
a, b = 1, 3

x = np.linspace(a, b, 400)
plt.figure(figsize=(10,6))

plt.plot(x, f(x), label='f(x)', lw=2)

plt.scatter([a, b], [f(a), f(b)], color='black')
plt.title('Extreme Value Theorem')
plt.legend()
plt.grid(True)
plt.show()
```

## The Intermediate Value Theorem

If $y = f(x)$ is a function that is defined and continuous on a closed interval $[a, b]$, then this function assumes every value between $f(a)$ and $f(b)$; that is, if $K$ is any number strictly between $f(a)$ and $f(b)$, then there exists at least one point $c$ in $(a, b)$ such that $f(c) = K$.
