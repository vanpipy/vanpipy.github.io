# Distance Formula

> A major fact of particular importance is Pythagorean theorem: In any right triangle, the sum of the squares of the legs equals the square of the hypotenuse.


```python
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
import math

w = 1
angle = 45
x0 = 0.5
y0 = 0
iw = math.sqrt(x0*x0 + (w - x0)*(w - x0))

fig, ax = plt.subplots()
rect0 = Rectangle((0, 0), w, w, color='b')
rect1 = Rectangle((x0, y0), iw, iw, color='r', angle=angle, rotation_point='xy')
ax.add_patch(rect0)
ax.add_patch(rect1)
ax.text(0.2, 0.02, 'a', fontsize=12)
ax.text(0.97, 0.2, 'a', fontsize=12)
ax.text(0.02, 0.2, 'b', fontsize=12)
ax.text(0.8, 0.02, 'b', fontsize=12)
ax.text(0.25, 0.25, 'c', fontsize=12)
ax.text(0.75, 0.25, 'c', fontsize=12)
fig.tight_layout()
```

As above graph shows,

$$(a + b)^2 = c^2 + 4*1/2ab$$
$$a^2 + 2ab + b^2 = c^2 + 2ab$$
$$a^2 + b^2 = c^2$$


```python
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle

x0 = 0.5
y0 = 0.5
x1 = 1.5
y1 = 1.5
x2 = x1
y2 = y0
offset = 0.1

fig, ax = plt.subplots()
plt.xlim(0, 2)
plt.ylim(0, 2)
ax.text(x0, y0 - offset, 'P0(x0, y0)')
ax.text(x1, y1 - offset, 'P1(x1, y1)')
ax.text(x2, y2 - offset, 'P2(x2, y2)')
plt.plot([x0, x1], [y0, y1])
plt.plot([x0, x2], [y0, y2], linestyle='dashed', color="r")
plt.plot([x1, x2], [y1, y2], linestyle='dashed', color="r")
fig.tight_layout()
```

It is easy to find that $P_0P_1^2 = P_0P_2^2 + P_1P_2^2$, so the distance formula is

$$d^2 = |x1 - x0|^2 + |y1 - y0|^2$$

$$d = \sqrt{(x1 - x0)^2 + (y1 - y0)^2}$$
