# The Concept Of A Function

> The most important concept in all of mathematics is that of a function.

What is a function? Briefly -- and we expand on this below -- if $x$ and $y$ are two variables that are related in such a way that whenever a permissible numerical value is assigned to $x$, there is determined one and only one corresponding numerical value for $y$, then $y$ is called a function of $x$.

## Polynomial Functions

$$1, x, x^2, x^3, ..., x^n, ...$$

$$p(x) = a_0 + a_1x + a_2x^2 + a_3x^3 + ... + a_nx^n$$


```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 200)
y = x**2 - 6*x + 3

# Create figure and plot
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='$y = x^2 - 6x + 3$')

# Add plot decorations
plt.title('Cubic Function Demonstration')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend()
plt.show()
```

## Rational Functions

$$\frac{1}{x}, \frac{x}{x^2 + 1}, \frac{x + 2}{x - 2}$$


```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 100)
y = 1 / x

# Create figure and plot
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='$y = 1 / x$')

# Add plot decorations
plt.title('Cubic Function Demonstration')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 100)
y = x / (x**2 + 1)

# Create figure and plot
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='$y = x / (x^2 + 1)$')

# Add plot decorations
plt.title('Cubic Function Demonstration')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend()
plt.show()
```

## Algebraic Functions

$$y = \sqrt{x}, y = \sqrt[3]{x^2 + 1}, y = x + \sqrt[3]{x^2 + 1}$$

$$y = x^{1/2}, y = (x^2 + 1)^{1/3}, y = x + (x^2 + 1)^{1/3}$$


```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 100)
y = np.sqrt(x)

# Create figure and plot
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='$y = \sqrt{x}$')

# Add plot decorations
plt.title('Cubic Function Demonstration')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 20, 100)
y = np.power(x**2 + 1, 1/3)

# Create figure and plot
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='$y = \sqrt[3]{x^2 + 1}$')

# Add plot decorations
plt.title('Cubic Function Demonstration')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend()
plt.show()
```

## Transcendental Functions

> Any function that is not algebraic is called __transcendental__.

The transcendental functions studied in calculus are the __trigonometric__, __inverse trigonometric__, __exponential__, and __logarithm__ functions.

$$y = \sin x, y = \cos x, y = \tan x$$
$$y = \arcsin x, y = \arccos x, y = \arctan x$$

$$y = a^x$$
$$y = \log_a{x}$$

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, np.pi * 2, 100)
y = np.sin(x)

# Create figure and plot
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='$y = \sin{x}$')

# Add plot decorations
plt.title('Cubic Function Demonstration')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-np.pi * 2, np.pi * 2, 100)
y = np.arcsin(x)

# Create figure and plot
plt.figure(figsize=(8, 6))
plt.plot(x, y, label="$y = arcsin x$")

# Add plot decorations
plt.title('Cubic Function Demonstration')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend()
plt.show()
```
