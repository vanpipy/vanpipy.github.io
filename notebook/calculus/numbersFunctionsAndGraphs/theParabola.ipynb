# The Parabola

The definition we use for a parabola is the following:

It is the curve consisting of all points that are equally distant from a fixed point $F$(called the focus) and a fixed line $d$(called the directrix). The distance from a point to a line is always understood to mean the perpendicular distance.

The condition expressed in the definition is stated algebraically by the equation:

$$\sqrt{x^2 + (y - p)^2} = y + p$$

$$x^2 + (y - p)^2 = (y + p)^2$$

$$x^2 + y^2 - 2py + p^2 = y^2 + 2py + p^2$$

$$x^2 = 4py$$

```python
import numpy as np
import matplotlib.pyplot as plt

p = 1  # Focus distance
x = np.linspace(-5, 5, 100)
y = x**2 / (4*p)

# Create figure
plt.figure(figsize=(8, 8))

# Plot parabola
plt.plot(x, y, label=f'Parabola $x^2 = 4py$ (p={p})')

# Plot focus and directrix
plt.scatter(0, p, color='red', label=f'Focus (0, {p})')
plt.axhline(y=-p, color='green', linestyle='--', label='Directrix $y = -{p}$')

# Add sample point P
px = 3
py = px**2 / (4*p)
plt.scatter(px, py, color='purple', label=f'Point P ({px}, {py:.2f})')

# Draw connection lines
plt.plot([px, px], [py, -p], 'grey', linestyle='--', label='Distance to directrix')
plt.plot([px, 0], [py, p], 'orange', linestyle='--', label='Distance to focus')

# Annotate the points
plt.annotate(f'P', (px, py), textcoords="offset points", xytext=(0, 10), ha='center')
plt.annotate(f'p(0, {p})', (0, p), textcoords="offset points", xytext=(0, 10), ha='center')

# Formatting
plt.title('Parabola with Focus and Directrix')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.axis('equal')
plt.show()
```
