# Middle Formula

> It is often useful to know the coordinates of the midpoint of the segment joining two given distinct points.

```python
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle

x0 = 0.5
y0 = 0.5
x1 = 1.5
y1 = 1.5
x2 = x0 + 1/2 * (x1 - x0)
y2 = y0 + 1/2 * (y1 - y0)
offset = 0.1

fig, ax = plt.subplots()
plt.xlim(0, 2)
plt.ylim(0, 2)
ax.text(x0, y0 - offset, 'P_0(x_0, y_0)')
ax.text(x1, y1 - offset, 'P_1(x_1, y_1)')
ax.text(x2, y2 - offset, 'P_2(x_2, y_2)')
plt.plot([x0, x1], [y0, y1])
fig.tight_layout()
```

$$
x = 1/2(x0 + x1), y = 1/2(y0 + y1)
$$
