# Find the x-intercept of the tangent

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate x values from -10 to 10
x = np.linspace(-10, 10, 100)
y = x**2
# slope
sy = 2 * x
# tangent line at x=a
a = 4.0
slope = 2 * a
tangent_line = slope * (x - a) + a**2

# Create plot
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='y = x')

# Add labels and title
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Quadratic Function y = x')
plt.grid(True)
plt.legend()

# Add tangent line at x=2 (slope = 4)
plt.plot(x, tangent_line, '--', color='red', label=f'Tangent at x={a} (slope={slope})')
plt.scatter(a, a**2, color='red', zorder=5)

# Calculate axis intercepts
x_intercept = a - (a**2)/slope  # Solve 0 - a = slope*(x - a)
y_intercept = -slope*a + a**2   # When x=0: y - a = slope*(-a)

# Plot intercept points
plt.scatter(x_intercept, 0, color='blue', zorder=5, label=f'X-intercept ({x_intercept:.1f}, 0)')
plt.scatter(0, y_intercept, color='green', zorder=5, label=f'Y-intercept (0, {y_intercept:.1f})')

# Add annotations
plt.annotate(f'({a:.1f}, {a**2:.1f})', (a, a**2), textcoords="offset points", xytext=(0,10), ha='center')
plt.annotate(f'({x_intercept:.1f}, 0)', (x_intercept, 0), textcoords="offset points", xytext=(0,10), ha='center')
plt.annotate(f'(0, {y_intercept:.1f})', (0, y_intercept), textcoords="offset points", xytext=(10,-2), ha='left')
```

Question: $y = x^2$ and the tangent line has a x-intercept equals 2.

* the tangent function is $sy = 2x$
* when $x = a$, so the slope is $2a$ and the tangent line: $y - a^2 = 2a(x - a)$
* $y = 2ax - 2a^2 + a^2$ -> $y = 2ax - a^2$
* when $x = 2$, $y = 0$, so $0 = 2a \times 2 - a^2$
* $-(a^2 - 4a) = 0$ -> $-a(a - 4) = 0$ -> $a = 0$ or $a = 4$ -> $(0, 0^2)$, $(4, 4^2)$
* so there are two points $(0, 0)$ and $(4, 16)$ which are the x-intercepts of the tangent line.

Question: $y = x^2$ and a point $(x_0, y_0)$, then here are always has x-intercept $1/2 * x_0$

* the tangent function is $sy = 2x$
* when $x = x_0$, so the slope is $2x_0$ and the tangent line: $y - x_0^2 = 2x_0(x - x_0)$
* $y = 2x_0x - 2x_0^2 + x_0^2$ -> $y = 2x_0x - x_0^2$
* when $y = 0$, $0 = 2x_0x - x_0^2$
* $2x_0x = x_0^2$ -> $x = x_0^2 / 2x_0$ -> $x = x_0 / 2$
