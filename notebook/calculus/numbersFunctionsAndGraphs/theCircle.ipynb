# The Circle

The distance formula is often useful in finding the equation of a curve whose geometric definition depends on one or more distances.

And one of the simplest curves of this kind is a __circle__, which can be defined as the set of all points at a given distance(the radius) from a given point(the center). If the center is the point $(h, k)$ and the radius is the positive number $r$, and if $(x, y)$ is an arbitrary point on the circle, then the definition condition says that: $$\sqrt{(x - h)^2 + (y - k)^2} = r$$

It is convenient to eliminate the radical sign by squaring, which yields $$(x - h)^2 + (y - k)^2 = r^2$$

And when the $h = k = 0$, then $$x^2 + y^2 = r^2$$

## The angle is a right angle

1. $$x^2 + y^2 = r^2$$
2. $$x^2 - r^2 = -y^2$$
3. $$(x + r)(x - r) = -y^2$$
4. $$\frac{y^2}{(x + r)(x - r)} = -1$$
5. $$\frac{y}{x + r} \cdot \frac{y}{x - r} = -1$$

```python
import numpy as np
import matplotlib.pyplot as plt

theta = np.linspace(0, np.pi, 100)
x = np.cos(theta)
y = np.sin(theta)

x1 = np.cos(np.pi / 3)
y1 = np.sin(np.pi / 3)

plt.title('Unit Circle')
plt.xlabel('x')
plt.ylabel('y')
plt.axis('equal')
plt.grid(True)
plt.plot(x, y, color="blue")
plt.plot([-1, 1], [0, 0], color="blue")
plt.plot([-1, x1], [0, y1], linestyle="dashed")
plt.plot([1, x1], [0, y1], linestyle="dashed")
plt.annotate("(x, y)", (x1, y1), textcoords="offset points", xytext=(0, 0), ha='left')
plt.annotate("(-r, 0)", (-1, 0), textcoords="offset points", xytext=(0, -10), ha='center')
plt.annotate("(r, 0)", (1, 0), textcoords="offset points", xytext=(0, -10), ha='center')
plt.show()
```
