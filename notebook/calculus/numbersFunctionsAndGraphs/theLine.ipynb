## The Slope Of A Line

> Any nonvertical straight line has a number associated with it that specifies its direction, called its __slope__.

This is number is defined as follows

Choose any two distinct points on the line, say $P_1 = (x_1, y_1)$ and $P_2 = (x_2, y_2)$. Then the slope is denoted by $m$ and defined be the radio:

$$m = \frac{y_2 - y_1}{x_2 - x_1}$$

$$m = \frac{y_2 - y_1}{x_2 - x_1} = \frac{y_1 - y_2}{x_1 - x_2}$$

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y = x + 1

x1 = 2
y1 = 3
x2 = 7
y2 = 8

plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.grid(True)

plt.annotate(f'P1({x1}, {y1})', (x1, y1), textcoords="offset points", xytext=(0,0), ha='left')
plt.annotate(f'P2({x2}, {y2})', (x2, y2), textcoords="offset points", xytext=(0,0), ha='left')

plt.plot(x, y, label='y = x + 1')
plt.plot([x1, x2], [y1, y1], linestyle="dashed")
plt.plot([x2, x2], [y1, y2], linestyle="dashed")
```
